```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str
    
def greeting_node(state: AgentState) -> AgentState:
    """" A personalized complimenting greeting node. """
    state['message'] = "Hey " + state["message"] + ", You're doing an amazing job learning LangGraph!"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("compli", greeting_node)

graph.set_entry_point("compli")
graph.set_finish_point("compli")


app = graph.compile()
```

```python
from IPython.display import display, Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message": "Abdulbasit"})
```
